# VISUALIZACÓN DE LOS DATOS

Veremos en un grafico la distribucion de la columna longitud y el ancho de los sepalos y petalos por especie , lo graficaremos en un grafico de caja

```python
# Aca crearemos dos graficos de caja para sacar la longitud y el ancho del sepalo

sns.boxplot(x='Especies', y='Longitud_Sepalo', data=df)
plt.title('Distribución de Longitud del Sepalo por Especie')
plt.xlabel('Especie')
plt.ylabel('Longitud del Sepalo')
plt.show()
```

```python
sns.boxplot(x='Especies', y='Ancho_Sepalo', data=df)
plt.title('Distribución de Ancho del Sepalo por Especie')
plt.xlabel('Especie')
plt.ylabel('Ancho del Sepalo')
plt.show()
```

```python
# Aca crearemos dos graficos de caja para sacar la longitud y el ancho de los Petalos
sns.boxplot(x='Especies', y='Longitud_Petalo', data=df)
plt.title('Distribución de Longitud del Petalo por Especie')
plt.xlabel('Especie')
plt.ylabel('Longitud del Petalo')
plt.show()
```

```python
sns.boxplot(x='Especies', y='Ancho_Petalo', data=df)
plt.title('Distribución de Ancho del Petalo por Especie')
plt.xlabel('Especie')
plt.ylabel('Ancho del Petalo')
plt.show()
```

Veremos como se relacionan entre si las dimenciones de los sepalos y de los petalos

```python
# Creamos un scatter plot para ver la relacion entre la longitud y el ancho del sepalo
sns.scatterplot(x='Longitud_Sepalo', y='Ancho_Sepalo', data=df, hue='Especies')
plt.title('Relación entre la longitud y el ancho del sepalo')
plt.xlabel('Longitud del Sepalo')
plt.ylabel('Ancho del Sepalo')
plt.show()
```

```python
# Creamos un scatter plot para ver la relacion entre la longitud y el ancho del Petalo
sns.scatterplot(x='Longitud_Petalo', y='Ancho_Petalo', data=df, hue='Especies')
plt.title('Relación entre la longitud y el ancho del Petalo')
plt.xlabel('Longitud del Petalo')
plt.ylabel('Ancho del Petalo')
plt.show()
```

```python
# Seleccionar solo las columnas numéricas
df_numerico = df[['Longitud_Sepalo', 'Ancho_Sepalo', 'Longitud_Petalo', 'Ancho_Petalo']]

# Calcular la matriz de correlación
matriz_correlacion = df_numerico.corr()

# Ver la matriz
print(matriz_correlacion)

```

```python
# Graficar la matriz de correlación usando seaborn
plt.figure(figsize=(8, 6))  # Tamaño del gráfico
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)

# Mostrar el gráfico
plt.title('Matriz de Correlación')
plt.show()
```

Veremos el promedio de cada columna de longitud y ancho  osea la media del sepalo y petalo segun la especie y lo graficamos en un grafico de barras

```python
# 1. Calcular el tamaño promedio de los sépalos y pétalos por especie
promedio_dimensiones = df.groupby('Especies')[['Longitud_Sepalo', 'Ancho_Sepalo', 'Longitud_Petalo', 'Ancho_Petalo']].mean()

# 2. Graficar los resultados en un gráfico de barras
promedio_dimensiones.plot(kind='bar', figsize=(10, 6))

# 3. Añadir etiquetas y título al gráfico
plt.xlabel('Especie')
plt.ylabel('Tamaño Promedio (cm)')
plt.title('Tamaño Promedio de Sépalos y Pétalos por Especie')

# Mostrar el gráfico
plt.xticks(rotation=0)  # Opcional: para que las etiquetas del eje x no se roten
plt.show()
```
